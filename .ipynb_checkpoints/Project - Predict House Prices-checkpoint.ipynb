# Download Data

````python
import os
import tarfile
import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
````

# Create a test set
Creamos un test set (o validation set). Ponemos 20% de los datos (típicamente) a un lado. Es importante que esos datos sean siempre los mismos, y que cada vez que se ejecute el notebook no cambie el train set y el test set. Esto lo podemos conseguir poniendo random_state a un número concreto

````python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
````

````python
housing["median_income"].hist()
````

````python
#Hacemos los 5 estratos o categorías. Añadimos una columna con la categoría a la que pertenece
housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
housing.head()
````

````python
housing["income_cat"].value_counts()
````

````python
from sklearn.model_selection import StratifiedShuffleSplit

#StratifiedShuffleSplit es una clase, el método split devuelve los índices de las filas en train y test set y con 
# pd.loc[] hacemos un panda con esas filas
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
````

Ahora tenemos train_set y test_set generados aleatoriamente y strat_train_set y strat_test_set generados teniendo en cuenta que queremos representar las categorías de median_income como están en el dataset. Comparamos cuál de los dos métodos tiene más error

````python
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)
#Value_counts devuelve un panda con el index y la frecuencia con la que se repite, ordenado. AL dividirlo entre 
#la longitud lo tenemos en porcentaje

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100
````

Vemos que hay un error menor con stratified

````python
compare_props
````

### Custom Transformers
Con custom transformers podemos hacer clases que heredan de BaseEstimator y de TransformerMixin con los métodos fit y transform, y de esta manera podemos usar unos métodos que directamente hacen las transformaciones que deseamos. En este caso hacemos un Transformer que genere los nuevos atributos a partir de los que teníamos inicialmente.

El atributo bedrooms_per_room se puede añadir o no (es un hyperparámetro). Ponerlo como parámetro permite luego ver si añadirlo o no mejora a los algoritmos de predicción, cuantos más hiperparámetros, más combinaciones podemos probar luego para encontrar la que mejor resultados nos de

````python
from sklearn.base import BaseEstimator, TransformerMixin

# column index
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
````

````python
housing_extra_attribs = pd.DataFrame(
    housing_extra_attribs,
    columns=list(housing.columns)+["rooms_per_household", "population_per_household"],
    index=housing.index)
housing_extra_attribs.head()
````

### Pipeline
En los datos numéricos hemos hecho las siguientes cosas:
+ Añadir nuevos atributos
+ Sustituir datos que faltan
+ Feature Scaling

En los datos categóricos hemos hecho:
+ One-hot Encoding

Ahora hacemos un Pipeline que agrupe todo esto

````python
housing.head() #Antes de hacer las transformaciones
````

````python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
    ])

housing_num_tr = num_pipeline.fit_transform(housing_num)
````

````python
housing_num_tr[0]
````

````python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num) #list devuelve los nombres de las columnas de un panda dataframe
cat_attribs = ["ocean_proximity"]

#ColumnTransformer aplica cierta función a las columnas y luego las junta
full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(handle_unknown='ignore'), cat_attribs),
    ])

housing_prepared = full_pipeline.fit_transform(housing)
````

````python
print(type(housing_prepared))
housing_prepared
 #Aquí ya tenemos los datos tal y como se los vamos a pasar al algoritmo de Machine Learning. Se lo pasamos
 # como un array (admite cualquier objeto array-like)
````

#### Compare models

````python

def compare_models (models, names, data_prepared, data_labels):
    compare_props = pd.DataFrame({
        "Model":[],
        "RMSE": [],
        "MAE": [],
        "RMSE Mean":[] ,
        "RMSE Std_Dev":[],
    }).sort_index()
    index=0
    for model in models:
        data_predictions = model.predict(data_prepared)
        mse = mean_squared_error(data_labels, data_predictions)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(data_labels, data_predictions)
        scores = cross_val_score(model, data_prepared, data_labels,
                             scoring="neg_mean_squared_error", cv=5)
        rmse_scores = np.sqrt(-scores)
        mean = rmse_scores.mean()
        std = rmse_scores.std()
        compare_props.loc[len(compare_props)] = ([names[index],rmse,mae,mean,std])
        index+=1
        
    
    return compare_props

compare_models([lin_reg, tree_reg, forest_reg,svm_reg_lin,svm_reg_rbf],['Linear Regression', 'Decision Tree Regressor','Random Forest Regressor', 'SVM Lineal', 'SVM Rbf'], housing_prepared, housing_labels)
````

# Fine-tune your model
Ahora que sabemos que el mejor de los modelos probados es RandomForestRegressor, intentamos encontrar los mejores parámetros de este con GridSearch y luego con Randomized Search

#### Grid Search

````python
from sklearn.model_selection import GridSearchCV

param_grid = [
    # try 12 (3×4) combinations of hyperparameters
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    # then try 6 (2×3) combinations with bootstrap set as False
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

forest_reg = RandomForestRegressor(random_state=42)
# train across 5 folds, that's a total of (12+6)*5=90 rounds of training 
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(housing_prepared, housing_labels)
````

````python
grid_search.best_params_
````

Por tanto, el mejor modelo es RandomForestRegressor con los siguientes parámetros

````python
grid_search.best_estimator_
````

#### Randomized Search

````python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=8),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(housing_prepared, housing_labels)
````

````python
rnd_search.best_estimator_
````

````python
compare_models([grid_search.best_estimator_,rnd_search.best_estimator_],['Random Forest Regressor Grid Search','Random Forest Regressor Randomized Search'], housing_prepared, housing_labels)
````

# Evaluate System on Test set
Ahora que estamos con el test set, solo usamos transform ya que fit se usa para aprender los parámetros. Si usáramos fit_transform aprendería la media y desviación de estos datos, y lo que queremos es que use la media y desviación del traning set para aplicarlos al test_set

````python
final_model = rnd_search.best_estimator_

X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"].copy()

X_test_prepared = full_pipeline.transform(X_test)
X_test_prepared = np.hstack((X_test_prepared,np.zeros((len(X_test_prepared),2)))) #Chapuza

final_predictions = final_model.predict(X_test_prepared)

final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
````

````python
final_rmse 
````

````python
#Intervalo de confianza del 95% de nuestro error:
from scipy import stats

confidence = 0.95
squared_errors = (final_predictions - y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(),
                         scale=stats.sem(squared_errors)))

    
````
